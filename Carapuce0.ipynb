### Preprocessing

```python
# Importation des packages
from sklearn.model_selection import train_test_split
```

```python
print(df['OnlineSecurity'].value_counts())
```

```python
trainset, testset = train_test_split(df, test_size=0.2, random_state=0)
```

```python
print(trainset['OnlineSecurity'].value_counts())
```

```python
# nettoyage
for col in df.select_dtypes('object'):
    print(f'{col: <50}{df[col].unique()}')
```

```python
def encodage(df):
    # Mapping pour les colonnes binaires 'Yes'/'No'
    binary_cols = ['Partner', 'Dependents', 'PhoneService', 'PaperlessBilling', 'Churn']
    mapping_binary = {'Yes': 1, 'No': 0}
    
    for col in binary_cols:
        df[col] = df[col].map(mapping_binary)
    
    # Mapping pour les colonnes avec des valeurs spcifiques
    specific_mapping = {
        'MultipleLines': {'Yes': 1, 'No': 0, 'No phone service': 0},
        'InternetService': {'DSL': 1, 'Fiber optic': 2, 'No': 0},
        'OnlineSecurity': {'Yes': 1, 'No': 0, 'No internet service': 0},
        'OnlineBackup': {'Yes': 1, 'No': 0, 'No internet service': 0},
        'DeviceProtection': {'Yes': 1, 'No': 0, 'No internet service': 0},
        'TechSupport': {'Yes': 1, 'No': 0, 'No internet service': 0},
        'StreamingTV': {'Yes': 1, 'No': 0, 'No internet service': 0},
        'StreamingMovies': {'Yes': 1, 'No': 0, 'No internet service': 0}
    }
    
    for col, mapping in specific_mapping.items():
        df[col] = df[col].map(mapping)
    
    # Encodage pour la colonne 'PaymentMethod'
    payment_mapping = {
        'Electronic check': 1,
        'Mailed check': 2,
        'Bank transfer (automatic)': 3,
        'Credit card (automatic)': 4
    }
    df['PaymentMethod'] = df['PaymentMethod'].map(payment_mapping)
    
    # Encodage pour la colonne 'Contract'
    contract_mapping = {
        'Month-to-month': 1,
        'One year': 2,
        'Two year': 3
    }
    df['Contract'] = df['Contract'].map(contract_mapping)
    
    return df
```

```python
# Elimination des valeurs manquantes
def imputation(df):
    return df.dropna(axis = 0)
```

```python
def preprocessing(df):
    df = encodage(df)
    df = imputation(df)
    X = df.drop('Churn',axis = 1) 
    y = df['Churn'] 
    print(y.value_counts())
    return X, y
```

```python
X_train,y_train = preprocessing(trainset)
```

```python
X_test,y_test = preprocessing(testset)
```

```python
print(X_test)
```

```python
# Evaluation du modele 
def evaluation(model):
    model.fit(X_train,y_train)
    ypred = model.predict(X_test)
    print(confusion_matrix(y_test,ypred))
    print(classification_report(y_test,ypred))
    
    N,train_score,val_score = learning_curve(model,X_train,y_train,cv = 4,train_sizes = np.linspace(0.1,1,10),scoring = 'f1')
    plt.figure(figsize=(12,8))
    plt.plot(N,train_score.mean(axis=1),label = 'train_score')
    plt.plot(N,val_score.mean(axis=1),label = 'validation score')
    plt.legend()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import make_pipeline
from sklearn.feature_selection import SelectKBest,f_classif
from sklearn.preprocessing import PolynomialFeatures,StandardScaler
```

```python
preprocessor = make_pipeline(PolynomialFeatures(2,include_bias = False),SelectKBest(f_classif,k = 10))
```

```python
RandomForest= make_pipeline(preprocessor,RandomForestClassifier(random_state = 0))
```

```python
AdaBoost = make_pipeline(preprocessor,AdaBoostClassifier(random_state = 0))
```

```python
SVM = make_pipeline(preprocessor,StandardScaler(),SVC(random_state = 0))
```

```python
KNN = make_pipeline(preprocessor,StandardScaler(),KNeighborsClassifier())
```

```python
list_of_models = [RandomForest,AdaBoost,SVM,KNN]
```

```python
# Evaluation du modele 
def evaluation(model):
    model.fit(X_train,y_train)
    ypred = model.predict(X_test)
    print(confusion_matrix(y_test,ypred))
    print(classification_report(y_test,ypred))
    
    N,train_score,val_score = learning_curve(model,X_train,y_train,cv = 4,train_sizes = np.linspace(0.1,1,10),scoring = 'f1')
    plt.figure(figsize=(12,8))
    plt.plot(N,train_score.mean(axis=1),label = 'train_score')
    plt.plot(N,val_score.mean(axis=1),label = 'validation score')
    plt.legend()
```

```python
def model_final(model, X, threshold=0):
    return model.decision_function(X) > threshold
```
